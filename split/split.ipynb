```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 加载原始图像（彩色图像）
img_path = "shengwu.png"  # 替换为你本地的图像路径
original = cv2.imread(img_path)
# import os
# print(os.path.exists(img_path))  # 如果返回 False，说明路径有问题
# 转为灰度图像


gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

# 应用二值化阈值（Thresholding）
# 127 是阈值，255 是最大值，返回两个值：ret是阈值，thresh是结果图像
ret, thresh_img = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# 显示结果
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Original Gray Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Thresholded Image (t={ret})")
plt.imshow(thresh_img, cmap='gray')
plt.axis('off')
plt.show()
```
